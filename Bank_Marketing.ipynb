#### Dados do Cliente:

1. age (idade) (numérico) <br>
2. job (emprego): tipo de trabalho (categórico: "administrador.","operário","empreendedor","empregada doméstica","gestão","aposentado","autônomo","serviços","estudante" ,"técnico","desempregado","desconhecido") <br>
3. marital: (conjugal) estado civil (categórico: "divorciado","casado","solteiro","desconhecido"; obs: "divorciado" significa divorciado ou viúvo) <br>
4. education: (educação) escolaridade (categórica: "básico.4a","básico.6a","básico.9a","ensino médio","analfabeto","curso.profissional","grau.universitário","desconhecido") <br>
5. default: (inadimplência) tem crédito inadimplente? (categórico: "não","sim","desconhecido") <br>
6. housing: (habitação) tem crédito habitação? (categórico: "não","sim","desconhecido") <br>
7. loan: (empréstimo) tem empréstimo pessoal? (categórico: "não","sim","desconhecido") <br>

#### Relacionado ao Último Contato da Campanha Atual:

8. contact: (contato) tipo de comunicação do contato (categórica: "celular","telefone") <br>
9. month: (mês) último mês de contato do ano (categórico: "jan", "fev", "mar", ..., "nov", "dez") <br>
10. day_of_week: (dia da semana) último dia de contato da semana (categórico: "seg","ter","qua","qui","sex") <br>
11. duration: (duração) duração do último contato, em segundos (numérico). Nota importante: este atributo afeta altamente o destino de saída (por exemplo, se duração=0 então y="não"). No entanto, a duração não é conhecida antes de uma chamada ser realizada. Além disso, após o término da chamada, y é obviamente conhecido. Assim, esta informação só deve ser incluída para fins de benchmark e deve ser descartada se a intenção for ter um modelo preditivo realista. <br>

#### Outros Atributos:

12. campaign: (campanha) quantidade de contatos realizados durante esta campanha e para este cliente (numérico, inclui último contato) <br>
13. pdays: número de dias que se passaram desde que o cliente foi contatado pela última vez em uma campanha anterior (numérico; 999 significa que o cliente não foi contatado anteriormente) <br>
14. previous: (anterior) quantidade de contatos realizados antes desta campanha e para este cliente (numérico) <br>
15. poutcome: resultado da campanha de marketing anterior (categórica: "fracasso","inexistente","sucesso") <br>

#### Atributos do Contexto  Social e Econômico:

16. emp.var.rate: taxa de variação do emprego - indicador trimestral (numérico) <br>  
17. cons.price.idx: índice de preços ao consumidor - indicador mensal (numérico) <br>  
18. cons.conf.idx: índice de confiança do consumidor - indicador mensal (numérico) <br>  
19. euribor3m: taxa euribor 3 meses - indicador diário (numérico) <br>  
20. nr.employed: número de empregados - indicador trimestral (numérico) <br>

#### Variável de Saída (variável target):

21. y - o cliente subscreveu um depósito a prazo? (binário: "sim","não") <br>

```python
def num_plot(df, col, title, symb):
    fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8,5),gridspec_kw={"height_ratios": (.2, .8)})
    ax[0].set_title(title,fontsize=18)
    sns.boxplot(x=col, data=df, ax=ax[0],color = "#1D8EF5")
    ax[0].set(yticks=[])
    sns.histplot(x=col, data=df, ax=ax[1],color = "#33AAFF", edgecolor="#1D1EA2")
    ax[1].set_xlabel(col, fontsize=16)
    plt.axvline(df[col].mean(), color='darkgreen', linestyle='--',linewidth=2.2, label='média=' + str(np.round(df[col].mean(),1)) + symb)
    plt.axvline(df[col].median(), color='red', linestyle='--',linewidth=2.2, label='mediana='+ str(np.round(df[col].median(),1)) + symb)
    plt.axvline(df[col].mode()[0], color='purple', linestyle='--',linewidth=2.2, label='moda='+ str(np.round(df[col].mode()[0],1)) + symb)
    plt.legend(bbox_to_anchor=(1, 1.03), ncol=1, fontsize=17, fancybox=True, shadow=True, frameon=True)
    plt.tight_layout()
    plt.show()
#função importada do Kaggle - How to make a successful Marketing Campaign? Autor: LUDOVICO CUOGHI 
```

# Tratamento dos Dados:

### De acordo com o prórpio banco de dados, a sugestão é excluir a coluna 'duração' da análise de dados!

```python
df=df.drop('duration', axis=1)
```

### Vamos substituir o valor 'unknown' das variáveis categóricas por 'NAN', para conseguirmos realizar o tratamente adequado:

```python
variaveis_categoricas
```

```python
df_clean = df.copy()
df_clean['job'] = df_clean['job'].replace('unknown', np.nan)
df_clean['education'] = df_clean['education'].replace('unknown', np.nan)
df_clean['marital'] = df_clean['marital'].replace('unknown', np.nan)
df_clean['default'] = df_clean['default'].replace('unknown', np.nan)
df_clean['housing'] = df_clean['housing'].replace('unknown', np.nan)
df_clean['loan'] = df_clean['loan'].replace('unknown', np.nan)
df_clean['contact'] = df_clean['contact'].replace('unknown', np.nan)
df_clean['month'] = df_clean['month'].replace('unknown', np.nan)
df_clean['day_of_week'] = df_clean['day_of_week'].replace('unknown', np.nan)
df_clean['poutcome'] = df_clean['poutcome'].replace('unknown', np.nan)
```

```python
df_clean.head(10)
```

```python
df_clean.isna().sum()
```

### Para ajustar a coluna target 'Y', que demonstra se o cliente aderiu ou não ao depósito, transformaremos os dados em uma dummy:

```python
df_clean = pd.get_dummies(df_clean, columns=['y'], drop_first=True)
df_clean.head(10)
```

```python
df_clean = df_clean.rename(columns={'y_yes': 'y'})
```

# Aplicando o Pycaret no Projeto: 

- PyCaret é uma biblioteca de machine learn de código aberto e low code em Python que automatiza fluxos de trabalho de aprendizado de máquina. É uma ferramenta completa de machine learn e gerenciamento de modelos que acelera exponencialmente o ciclo do experimento e torna você mais produtivo. <br>
- Em comparação com outras bibliotecas de machine learn de código aberto, PyCaret é uma biblioteca alternativa de low code que pode ser usada para substituir centenas de linhas de código por apenas algumas palavras. Isso torna os experimentos exponencialmente rápidos e eficientes. PyCaret é essencialmente um wrapper Python em torno de várias bibliotecas e estruturas de machine learn, como scikit-learn, XGBoost, LightGBM, CatBoost, Optuna, Hyperopt, Ray e muitos mais. <br>

#### Indicando as configurações preliminares do Setup do Pycaret: <br>
    - o dataframe definido ('data');

    - permitir um certo balanceamento da distribuição das variáveis analisadas do dataset;
